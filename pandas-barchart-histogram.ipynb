```python
import pandas as pd    # for working with "rows/columns" oriented data
import urllib.request  # for loading documents using http

%matplotlib inline

import matplotlib # plotting
import matplotlib.pyplot as plt

# for data see http://github.com/sfsheath/roman-amphitheaters
# note: province and augustan region data entry very much underway so
# the graphs are incomplete. they are not reliable indicators of the
# distribution of amphitheaters around the empire.
# except that it's not wrong to note that there are many in regio i.


# load csv file
ramphs = pd.read_csv("https://raw.githubusercontent.com/sfsheath/roman-amphitheaters/master/roman-amphitheaters.csv")
pd.__version__
```

```python
provinces = ramphs['province'].dropna()
regions   = ramphs['italyregion'].dropna()
```

```python
provinces.groupby(provinces).count().sort_values().plot.bar()
```

```python
regions.groupby(regions).count().sort_values().plot.bar()
```

```python
# "bar chart" counting amphitheaters by region
combined = pd.concat((provinces,regions))

combined.groupby(combined).count().sort_values().plot.bar()
```

```python
# histogram showing distribution of amphitheater longitudes (that is, west to east).
ramphs['longitude'].plot.hist()
```
