# Load from Google, Combine colums

Quick demo of loading from a Google Sheet that has been 'published to web'. Once loaded, pd.concat() is used to combine columns that hold similar information.

DATA IS VERY INCOMPLETE. See http://github.com/sfsheath/roman-amphitheaters for more info.

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import pandas as pd

# load amphitheater csv file from a google sheet

ramphs = pd.read_csv("https://docs.google.com/spreadsheets/d/1XRRe-74Gxb-YErNldbPP9VW1IARH-JsWXIHMLnyZhfc/pub?output=csv")

# you can see the data at:
# https://docs.google.com/spreadsheets/d/1XRRe-74Gxb-YErNldbPP9VW1IARH-JsWXIHMLnyZhfc/edit#gid=0
```

```python
# all good?
ramphs.head(2)
```

```python
# province and italyregion hold similar information
# create two pandas.Series variables ...
provinces    = ramphs['province'].dropna()
italyregions = ramphs['italyregion'].dropna()
# in the above, .dropna() gets rid of incomplete data. Very useful.

# ... then combine them
regions = pd.concat((provinces,italyregions))

regions

# note that in output the 'index' isn't renumbered. Look at end.
```

```python
# with renumbering of index
regions = pd.concat((provinces,italyregions), ignore_index = True)
regions
```

```python
# how many per 'generic' region
regions.groupby(regions).count()
```

```python
# barchart with sorting
regions.groupby(regions).count().sort_values().plot.bar()
```
